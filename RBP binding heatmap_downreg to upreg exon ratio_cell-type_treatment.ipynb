```
# Fixing random seed to 007 for reproducibility
from numpy.random import seed
seed(7)
from tensorflow import set_random_seed
set_random_seed(7)
```

```
# Importing essentials
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Load data
dataset = pd.read_csv('SE_RBP analysis_B_M_T_EtOH_GDM_Seaborn_input.csv')
```

```
# Display all data
dataset.head(100)
```

```
# Describe the data; 56 samples, range of 0 to 1 (not z-scored)
dataset.describe()
```

```
# Set Seaborn Style defaults
sns.set(style = 'ticks', font= 'sans-serif', font_scale=1.25)
```

```
#Set up Matplotlib figure
f,ax = plt.subplots(figsize=(150, 75))

# Specify colors
cmap = sns.diverging_palette(240, 10, as_cmap=True)

# Plotting the heatmap with the correct aspect ratio
sns.heatmap(dataset, linewidths=1, linecolor = 'white', cmap=cmap, center=1, square=True, vmin = 0, vmax = 2, cbar_kws={'shrink':1}, annot = True, fmt='0.2g' )

#Save figure as svg
plt.savefig ('SE_RBP analysis_B_M_T_EtOH_GDM_heatmap.svg', format = 'svg')
```

```
#No annotations inside cells

#Set up Matplotlib figure
f,ax = plt.subplots(figsize=(50,25))

# Specify colors
cmap = sns.diverging_palette(240, 10, as_cmap=True)

# Plotting the heatmap with the correct aspect ratio
sns.heatmap(dataset, linewidths=0.5, linecolor = 'white', cmap=cmap, center=1, vmin = 0, vmax = 2, square=True, cbar_kws={'shrink':1})

#Save figure as svg
plt.savefig ('SE_RBP analysis_B_M_T_EtOH_GDM_heatmap_no annot.svg', format = 'svg')
```
