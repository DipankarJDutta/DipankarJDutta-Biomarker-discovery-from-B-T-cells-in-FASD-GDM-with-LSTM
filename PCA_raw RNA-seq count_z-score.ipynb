```
# Fix random seed to 007 for reproducibility
from numpy.random import seed
seed(7)
from tensorflow import set_random_seed
set_random_seed(7)
```

```
# Load essentials
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```
# Load data
df = pd.read_csv('JS count_raw HTseq count_NEW.csv', delimiter = ',')
```

```
# Display data top
df.head()
```

```
# Display data bottom
df.tail()
```

```
# Z-score data
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
print(scaled_df)
```

```
# Export Z-scored data to csv
np.savetxt('JS count_raw HTseq count_NEW_Z-score.csv', scaled_df, fmt ='%.3f', delimiter = ',')
```
