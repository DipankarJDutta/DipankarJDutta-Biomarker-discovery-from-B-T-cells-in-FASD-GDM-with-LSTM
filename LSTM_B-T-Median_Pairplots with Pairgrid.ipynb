```
# Fixing random seed to 007 for reproducibility
from numpy.random import seed
seed(7)
from tensorflow import set_random_seed
set_random_seed(7)
```

```
# Importing essentials
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Load data
dataset = pd.read_csv('B-T-median_+col-names_+learner-type.csv')
```

```
# Display top of data
dataset.head()
```

```
# Describe the data; 56 samples, range of 0 to 1 (not z-scored)
dataset.describe()
```

```
# Set Seaborn Style
sns.set(style = 'ticks', font= 'sans-serif', font_scale=1.25)

# Plot Pairplot
sns.pairplot(dataset, corner=True, kind="reg", hue='Learner')

#Save figure as svg
plt.savefig ('pairplot_B-T_median_diag-hist.svg', format = 'svg')
```
