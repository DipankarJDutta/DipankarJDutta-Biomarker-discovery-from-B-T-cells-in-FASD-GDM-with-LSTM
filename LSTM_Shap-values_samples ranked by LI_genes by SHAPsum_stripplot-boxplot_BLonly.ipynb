```
# Fixing random seed to 007 for reproducibility
from numpy.random import seed
seed(7)
from tensorflow import set_random_seed
set_random_seed(7)
```

```
# Importing essentials
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Load data
dataset = pd.read_csv('Shap values_LSTM_Stripplot-input - BLonly.csv')
```

```
# Display all data
dataset.head(56)
```

```
# Describe the data; 56 samples, range of 0 to 1 (not z-scored)
dataset.describe()
```

```
# Set Seaborn Style defaults

# White background; font size and ratio
sns.set(style = 'white', font= 'sans-serif', font_scale=1.25)

# Mark ticks on the X and Y axis
sns.set_style("ticks")
```

```
# Strip plot:
#Set up Matplotlib figure
f,ax = plt.subplots(figsize=(60,30))

# Create an array with the green and red colors
colors = ["#ff0000"]

# Set your custom color palette
sns.set_palette(sns.color_palette(colors))

# Plotting the heatmap with the correct aspect ratio
sns.stripplot(x='Shap-values', y='Genes', hue='Learner', data=dataset, linewidth=0.5, dodge=False, size = 10, edgecolor = 'black', alpha=0.50 )

#Save figure as svg
plt.savefig ('Strip-plot_shap_LSTM_BL.svg', format = 'svg')
```

```
#Set up boxplot on top of dodged strip-plot

#Set up Matplotlib figure
f,ax = plt.subplots(figsize=(75,37.5))
ax = sns.boxplot(x='Shap-values', y='Genes', data=dataset, order = ('2310001H17Rik', 'Kdm7a', 'Mtpap', 'Ets1-T', 'Tvp23b', 'Ttc3', 'Wdr43', 'Ptcd3', 'Pld4', 'Rars2', 'Dapp1', 'Chchd7', 'Ets1', 'Mapk9', 'Ddx50', 'Tnfaip3', 'Brox', 'Tcrg-C4', 'Rasa1', 'Gnas', 'Bin1', 'Umps', 'Zfp639', 'Ms4a6b', 'Stk38', 'Wac', 'Caelf2', 'Setdb2', 'Usp15'), whis=np.inf, color='#e7e7e7', linewidth=0.5)
ax = sns.stripplot(x='Shap-values', y='Genes',  hue='Learner', data=dataset, order = ('2310001H17Rik', 'Kdm7a', 'Mtpap', 'Ets1-T', 'Tvp23b', 'Ttc3', 'Wdr43', 'Ptcd3', 'Pld4', 'Rars2', 'Dapp1', 'Chchd7', 'Ets1', 'Mapk9', 'Ddx50', 'Tnfaip3', 'Brox', 'Tcrg-C4', 'Rasa1', 'Gnas', 'Bin1', 'Umps', 'Zfp639', 'Ms4a6b', 'Stk38', 'Wac', 'Caelf2', 'Setdb2', 'Usp15'), linewidth=0.25, dodge=True, size = 25, edgecolor = 'black', alpha=0.50, jitter=True)

# Set axes limits
ax.set(xlim=(-0.085, 0.085))

# Remove top, right and left borders
sns.despine(left=True)

# Decrease line-widths ox X and y axes
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(0.5)

#Save figure as svg
plt.savefig ('Strip-plot_box-plot_shap_LSTM_BL.svg', format = 'svg')
```
