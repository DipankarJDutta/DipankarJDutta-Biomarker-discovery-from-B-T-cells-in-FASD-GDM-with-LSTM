```
# Load data
dataset = pd.read_csv('B-T-median_+col-names_-learner-type.csv')
```

```
# Importing essentials
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Set Seaborn Style
sns.set(style = "white")
```

```
# Pearson Correlation of columns
corr = dataset.corr(method='pearson')

#Set up Matplotlib figure
f,ax = plt.subplots(figsize=(20,15))

# Mask the mirror image across the diagonal
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Specify colors
cmap = sns.diverging_palette(240, 10, as_cmap=True)

# Plotting the heatmap without the mask and correct aspect ratio
sns.heatmap(corr, linewidths=.5, cmap=cmap, center=0, vmax=1, square=True, cbar_kws={'shrink':0.5})

```
