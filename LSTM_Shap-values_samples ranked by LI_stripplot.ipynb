```
# Fixing random seed to 007 for reproducibility
from numpy.random import seed
seed(7)
from tensorflow import set_random_seed
set_random_seed(7)
```

```
# Importing essentials
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Load data
dataset = pd.read_csv('Shap values_LSTM_Stripplot-input.csv')
```

```
# Display all data
dataset.head(56)
```

```
# Describe the data; 56 samples, range of 0 to 1 (not z-scored)
dataset.describe()
```

```
# Set Seaborn Style defaults
sns.set(style = 'whitegrid', font= 'sans-serif', font_scale=1.25)
```

```
#Set up Matplotlib figure
f,ax = plt.subplots(figsize=(25, 12.5))

# Specify colors
cmap = sns.diverging_palette(240, 10, as_cmap=True)

# Plotting the heatmap with the correct aspect ratio
sns.stripplot(x='Shap-values', y='Genes', hue='Learner', data=dataset)

#Save figure as svg
plt.savefig ('Strip-plot_shap_LSTM_ranked by LI.svg', format = 'svg')
```
